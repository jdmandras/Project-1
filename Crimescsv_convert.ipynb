```python
import pandas as pd
from pathlib import Path
```

```python
#File to Load
crimes_csv = Path("./Resources/Crimes_-_2001_to_Present.csv")

#Read Data FIle and Store into Pandas DataFrames
crimes_data = pd.read_csv(crimes_csv)

#Display Data Frame (ID,Case Number,Date,Block,IUCR,Primary Type,Description,Location Description,Arrest,Domestic,Beat,District,Ward,Community Area,FBI Code,X Coordinate,Y Coordinate,Year,Updated On,Latitude,Longitude,Location)
crimes_data.head()
```

```python
#Display Data Frame (Date, IUCR, Primary Type, Description, Location Description, Arrest, District, FBI Code, Year, Latitude, Longitude)
crime_data = crimes_data[["Date","IUCR","Primary Type", "Description", "Location Description", "Arrest", "District", "FBI Code", "Year", "Latitude", "Longitude"]]
crime_data
```

```python
crime_data.astype({"Arrest": str}).dtypes
crime_data.dtypes
```

```python
#Drop Rows where Year <2015
crime_data_filteredyears = crime_data[crime_data["Year"] >= 2015]

#Drop Rows where Arrest = False
arrests_df = crime_data_filteredyears[crime_data_filteredyears["Arrest"] == True]

#Another Resource for no arrests
noarrests_df = crime_data_filteredyears[crime_data_filteredyears["Arrest"] == False]
```

```python
crime_years = crime_data_filteredyears.sort_values(by=['District',"Primary Type", "Year"], ascending=True)
crime_years = crime_years.sort_values(by = ["Arrest"], ascending = False)
crime_years = crime_years.set_index(["District", "Primary Type", "Arrest"])

#crime_years = crime_years.set_
crime_years

crime_years.to_csv("Crimes_2015toPresent.csv")
```

```python
# Group by District and find the maximum number of arrests
highest_arrests_per_district = crime_data_filteredyears.groupby("District")["Primary Type"].sum().reset_index()

# Sort the DataFrame by the number of arrests in descending order
#highest_arrests_per_district = highest_arrests_per_district.sort_values(by='Arrest', ascending=False)

# Display DataFrame
highest_arrests_per_district
```
