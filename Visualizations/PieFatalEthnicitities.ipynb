```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import csv file
fatal = pd.read_csv("../Resources/Fatal_complete.csv")
fatal
```

```python
FRace = fatal["Race"].value_counts()
FRace
```

```python
FRace = pd.DataFrame(FRace)

FRace
```

```python
plt.gca().axis("equal")
pie = plt.pie(FRace["Race"], startangle=0, autopct='%1.0f%%', pctdistance=0.9, radius=1.2)
labels=FRace.index.unique()
plt.title('Percentage of Fatalities 2015-2023 by Race', weight='bold', size=14)
plt.legend(pie[0],labels, bbox_to_anchor=(1,0.5), loc="best", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
plt.subplots_adjust(left=0.0, bottom=0.1, right=0.85)

plt.show()
plt.clf()
plt.close()
```
