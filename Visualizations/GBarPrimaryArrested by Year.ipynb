```python
#Data Frame of Arrested
Ayear2015 = year2015[year2015["Arrest"] == True].groupby("Primary Type").size()
Ayear2015
```

```python
#Data Frame of Not Arrested
Nyear2015 = year2015[year2015["Arrest"] == False].groupby("Primary Type").size()
Nyear2015
```

```python

Year = ("2015", "2016", "2017", "2018", "2019", "2020")
primary_types = {
    'Homicide': (6, 4, 3, 11, 4, 0),
    'Theft': (4135, 6373, 6376, 6233, 2809, 180),
    'Battery': (6942, 10144, 9878, 10065, 4839, 554),
    'Human Trafficking': (1, 1, 1, 1, 0, 0)
  
}

x = np.arange(len(Year))  # the label locations
width = 0.1  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in primary_types.items():
    offset = width * multiplier
    rects = ax.bar(x+offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Arrests')
ax.set_title('Arrested : Two Most and Least Occurring Crime Types by Year')
ax.set_xticks(x + width, Year)
ax.legend(loc='best')


plt.show()
```
