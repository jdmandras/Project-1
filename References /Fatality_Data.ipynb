```python
import pandas as pd
from pathlib import Path
```

```python
#File to Load
fatalities_shoot_csv = Path("../References /Fatal_Shootings/fatalities_shoot.csv")
fatale_csv = Path("../References /Fatal_Encounters/fatale.csv")
#Read Data FIle and Store into Pandas DataFrames
fatalities_shoot_data = pd.read_csv(fatalities_shoot_csv)
fatale_data = pd.read_csv(fatale_csv)
fatalities_shoot_data["Date"] = fatalities_shoot_data["date"]
#merge Data Frames
fatal_data = pd.merge(fatale_data, fatalities_shoot_data, how="left", on=["Date"])
fatal_data.head()
```

```python
fatal_data = fatal_data[["Date", "Name","Age", "Gender", "Race",	"Location of injury (address)",	"Location of death (city)",	"State", "Location of death (zip code)", "Location of death (county)","Full Address","Latitude","Longitude", "location_precision","name", "age", "gender", "race", "body_camera"]]
fatal_data
```

```python
# Convert the date to datetime64
fatal_data["Date"] = pd.to_datetime(fatal_data["Date"], infer_datetime_format= "%m/%d/%y")
  
# Filter data between two dates
filtered_df = fatal_data.loc[(fatal_data["Date"] >= '2015/01/01')]
# Display
filtered_df = filtered_df.sort_values(by="Date", ascending = True)
filtered_df = filtered_df.set_index("Date")
filtered_df
```

```python
filtered_df.to_csv("Fatal_complete.csv")
```
