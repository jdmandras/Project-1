```python
import pandas as pd
import plotly.express as px

# Read the CSV file
crimes = pd.read_csv("Resources/Crimes_2015toPresent.csv")

# Filter the data for arrests in 2015 and remove null values
arrests_2015 = crimes[(crimes["Arrest"] == True) & (crimes["Year"] == 2015)].dropna(subset=["Latitude", "Longitude"])

# Create a scatter mapbox plot using Plotly Express
fig = px.scatter_mapbox(arrests_2015, lat="Latitude", lon="Longitude", zoom=10)

# Update the map layout
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(title="Arrests with Latitude and Longitude in 2015")
fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})

# Display the plot
fig.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
crimes = pd.read_csv("Resources/Crimes_2015toPresent.csv")

# Filter the data for arrests in 2015 and remove null values
arrests_2015 = crimes[(crimes["Arrest"] == True) & (crimes["Year"] == 2015)].dropna(subset=["Latitude", "Longitude"])

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(arrests_2015.groupby(["Latitude", "Longitude"]).size().unstack(fill_value=0), cmap="YlOrRd")
plt.title("Arrest Heatmap in 2015")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Display the plot
plt.show()
```
